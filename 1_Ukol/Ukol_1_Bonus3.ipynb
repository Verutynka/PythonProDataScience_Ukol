```python
import pandas as pd
import statsmodels.formula.api as smf

data = pd.read_csv("df.csv")
data
```

```python
OHE = pd.get_dummies(data["Sector"])
OHE
```

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT + OHE['Basic Materials'] + OHE['Communication Services'] + OHE['Conglomerates'] + OHE['Consumer Cyclical'] + OHE['Consumer Defensive'] + OHE['Energy'] + OHE['Financial Services'] + OHE['Healthcare'] + OHE['Industrials'] + OHE['Technology']"

mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```
