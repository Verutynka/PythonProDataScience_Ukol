- Bonus 2: Použij původní (tedy "nerobustní") model a vyčísli Cookovu vzdálenost pro všechny hodnoty. Vyřaď všechny akcie s Cookovou vzdáleností vyšší než 1 a přepočítej regresní model. Nakonec opět najdi potenciálně nejvíce podhodnocenou akcii a podívej se, jestli jde o stejnou akcii jako u předchozích modelů.

```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import OLSInfluence

data = pd.read_csv("df.csv")
data.head()
```

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"

mod = smf.ols(formula=formula, data=data)
res = mod.fit()
influence = OLSInfluence(res)
data["Cook_Distance"] = influence.cooks_distance[0]
data.head()
```

```python
#data_cook = data.sort_values(by="Cook_Distance", ascending=False)
#data_cook.head()
```

```python
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

```python
#vyřazení hodnot s Cookovou vzdáleností vyšší než 1

data = data[data["Cook_Distance"] < 1]
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

```python
data["Y_Predicted"] = res.predict(data)
data.head()
```

```python
data["Y_difference"] = data["Y_Predicted"] - data["Y"]
data.head()
```

```python
top_5_Y_differences = data.sort_values(by="Y_difference", ascending=False)
top_5_Y_differences.head()
```
